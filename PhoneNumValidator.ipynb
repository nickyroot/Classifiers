```python
#Import
telNums = read_csv('TelNums.csv')
```

```python
#Remove whiteSpace
def removeWhiteSpace (y):
    #y = y.strip()
    z = y.replace(" ", "")
    return z
```

```python
#Lets replace question marks and merge the two columns into 1 with no spaces
telNums['TEL_ARA_CD'] = telNums['TEL_ARA_CD'].str.replace('?',' ')
telNums["Merged"] = telNums["TEL_ARA_CD"] + telNums["TEL_NO"]
telNums["MergedNoSpace"] = telNums["Merged"].apply(removeWhiteSpace)
```

```python
def validateTelNum (x):
    rule = re.compile('^\s*\(?(020[78]\)? ?[1-9][0-9]{2} ?[0-9]{4})|(0[1-8][0-9]{3}\)? ?[1-9][0-9]{2} ?[0-9]{3})\s*$')
    chain = itertools.chain(*rule.findall(x))
    if x in list(chain):
        return 1
    else: 
        return 0
    
```

```python
validateTelNum('018845292')
```

```python
telNums["Validation"] = telNums["MergedNoSpace"].apply(validateTelNum)
```
