```python
%pylab inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#populating the interactive namspace from numpy and matplotlib
```

```python
beer = pd.read_csv('beer.txt', delimiter='\t')
```

```python
beer.head()
```

```python
#janitor work, lets get rid of the n/a's
beer = beer.dropna()
```

```python
beer.shape
```

```python
# for classifitation we need labels
def good(x):
    if x>4.3:
        return 1
    else:
        return 0

# apply the function to the WR column
beer['Good']=beer['WR'].apply(good)
```

```python
beer.head() # so you can call the function as you add a column!!!
```

```python
# create a logistic model instance
from sklearn import linear_model

logm = linear_model.LogisticRegression()
```

```python
my_input = beer[['Reviews','ABV']].values
good_output = beer['Good'].values
good_output
```

```python
# lets fit the model
logm.fit(my_input, good_output)
```

```python
logm.predict(my_input)
```

```python
logm.score(my_input, good_output)
```

```python
# lets change the penalty
logm2 = linear_model.LogisticRegression()
logm2.set_params(penalty='l2',dual=False)
```

```python
logm2.fit(my_input, good_output)
```

```python
logm2.predict(my_input)
```

```python
logm2.score(my_input, good_output)
```
